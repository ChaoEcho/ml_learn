## 决策树

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('D:/Resource/EchoChao/Projects/Pycharm/learing/other_learn/week5/iris_data.csv')
data.head()
```

```python
X = data.drop(['target','label'],axis=1)
y= data.loc[:,'label']
X.shape,y.shape
```

```python
from sklearn import tree
dc_tree = tree.DecisionTreeClassifier(criterion='entropy',min_samples_leaf=5)
dc_tree.fit(X,y)
```

```python
y_predict = dc_tree.predict(X)
from sklearn.metrics import accuracy_score
accuracy = accuracy_score(y,y_predict)
accuracy
```

```python
fig = plt.figure(figsize=(10,10))
tree.plot_tree(dc_tree,filled='True',feature_names=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'],class_names=['setosa','versicolor','virginica'])
```

```python
dc_tree = tree.DecisionTreeClassifier(criterion='entropy',min_samples_leaf=3)
dc_tree.fit(X,y)
```

```python
y_predict = dc_tree.predict(X)
```

```python
fig = plt.figure(figsize=(10,10))
tree.plot_tree(dc_tree,filled='True',feature_names=['SepalLength', 'SepalWidth', 'PetalLength', 'PetalWidth'],class_names=['setosa','versicolor','virginica'])
```

## 异常检测

```python
data = pd.read_csv('D:/Resource/EchoChao/Projects/Pycharm/learing/other_learn/week5/anomaly_data.csv')
data.head()
```

```python
fig1 = plt.figure(figsize=(10,5))
plt.scatter(data.loc[:,'x1'],data.loc[:,'x2'])
plt.title('data')
plt.xlabel('x1')
plt.ylabel('x2')
plt.show()
```

```python
x1 = data.loc[:,'x1']
x2 = data.loc[:,'x2']
```

```python
fig2 = plt.figure(figsize=(10,5))
plt.subplot(121)
plt.hist(x1,bins=100)
plt.title('x1 distribution')
plt.xlabel('x1')
plt.ylabel('counts')
plt.subplot(122)
plt.hist(x2,bins=100)
plt.title('x2 distribution')
plt.xlabel('x2')
plt.ylabel('counts')
plt.show()
```

```python
x1_mean = x1.mean()
x1_sigma = x1.std()
x2_mean = x2.mean()
x2_sigma = x2.std()
x1_mean,x1_sigma,x2_mean,x2_sigma
```

```python
from scipy.stats import norm
x1_range = np.linspace(0,20,300)
x1_normal = norm.pdf(x1_range,x1_mean,x1_sigma)
x2_range = np.linspace(0,20,300)
x2_normal = norm.pdf(x2_range,x2_mean,x2_sigma)
```

```python
fig3 = plt.figure(figsize=(20,5))
plt.subplot(121)
plt.plot(x1_range,x1_normal)
plt.title('normal p(x1)')
plt.subplot(122)
plt.plot(x2_range,x2_normal)
plt.title('normal p(x2)')
plt.show()
```

```python
from sklearn.covariance import EllipticEnvelope
ad_model = EllipticEnvelope(contamination=0.02)
ad_model.fit(data)
```

```python
y_predict = ad_model.predict(data)
print(pd.value_counts(y_predict))
```

```python
#visualize the result
fig4 = plt.figure(figsize=(10,6))
orginal_data=plt.scatter(data.loc[:,'x1'],data.loc[:,'x2'],marker='x')
anomaly_data=plt.scatter(data.loc[:,'x1'][y_predict==-1],data.loc[:,'x2'][y_predict==-1],marker='o',facecolor='none',edgecolor='red',s=150)

plt.xlabel('$x_1$')
plt.ylabel('$x_2$')
plt.legend((orginal_data,anomaly_data),('raw data','error data'))
plt.axis([4.5,15,2.5,15])
plt.show()
```
